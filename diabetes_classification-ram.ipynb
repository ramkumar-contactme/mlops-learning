# Fetch the data

```python
import pandas as pd
```

```python
df=pd.read_csv("diabetes.csv")
```

# Analyze the data

```python
df.info()
```

# Identify inputs and output

```python
y = df.Outcome # output, target
X = df.drop(columns=['Outcome']) # input, features
```

```python
X.head()
```

# Choose an ML algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

# Train the model

```python
model.fit(X,y)
```

# Evaluate the model

```python
model.score(X,y)
```

# Make predictions

```python
model.predict([[3.4,4.3,4.5,2.3]])
```
