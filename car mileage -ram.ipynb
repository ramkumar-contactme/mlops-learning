```python
import pandas as pd
```

```python
df = pd.read_csv("mpg.csv")
```

```python
df.info()
```

```python
y = df.MPG # output, target
X = df.drop(columns=[ 'MPG', 'Origin' ]) # input, features
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[6,199.0,	90,	2648,	15.0,	70]])
```
